### Score Distribution 

```python
plt.figure(figsize=(10,6))
sns.distplot(df["Score"])
plt.title("Score", size=15)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.distplot(df["Perceptions of corruption"])
plt.title("Perceptions of corruption", size=15)
plt.show()
```

### GDP Per Capita and Score 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["GDP per capita"], y=df["Score"])
plt.title("GDP per capita vs Score", size=15)
plt.show()
```

### Social Support and Score 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Social support"], y=df["Score"])
plt.title("Social Support vs Score", size=15)
plt.show()
```

### Life Expectancey And Score 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Healthy life expectancy"], y=df["Score"])
plt.title("Healthy life expectancy vs Score", size=15)
plt.show()
```

### Life Choices and Score 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Freedom to make life choices"], y=df["Score"])
plt.title("Freedom to make life choices vs Score", size=15)
plt.show()
```

### Generosity And Score 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Generosity"], y=df["Score"])
plt.title("Generosity vs Score", size=15)
plt.show()
```

### Perceptions Of Corruption and Score 

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df["Perceptions of corruption"], y=df["Score"])
plt.title("Perceptions of corruption vs Score", size=15)
plt.show()
```

### Top 10 Countries 

```python
happiest_10_countries = df.iloc[:10]
plt.figure(figsize=(12,6))
sns.barplot(x=happiest_10_countries["Country or region"], y=df["Score"])
plt.title("Scores of Happiest 10 Countries'", size=15)
plt.show()
```

### Bottom 10 Countries 

```python
Bottom_10_countries = df.iloc[-10:]
plt.figure(figsize=(12,6))
sns.barplot(x=Bottom_10_countries["Country or region"], y=df["Score"])
plt.xticks(rotation=30)
plt.show()
```
